```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from numpy import genfromtxt

plt.rcParams['axes.grid'] = False
plt.rcParams['grid.alpha'] = 1
plt.rcParams['grid.color'] = "#cccccc"
plt.rcParams.update({"font.size": 15, "font.family": "serif"})
```

```python
# Importamos los datos obtenidos en Mathematica (archivo .csv)
Q2 = genfromtxt('emptyTable25-10.csv', delimiter=',')
lineaarray = genfromtxt('linealist25-10.csv', delimiter=',')
```

```python
# Rangos de los mu's

mu3_0 = 0.1
mu4_0 = 0.1

mu3_f = 12.5664
mu4_f = 3.94784

# Punto crítico

mu3_crit = 8.71837
mu4_crit = 2.92986

# Línea transición

mu3_linea = lineaarray[:,0]
mu4_linea = lineaarray[:,1]

```

```python
fig, ax = plt.subplots(figsize=(20,10))

c = ax.imshow(Q2, cmap ='viridis', interpolation ='nearest', origin ='lower', extent =[mu3_0, mu3_f, mu4_0, mu4_f],aspect='auto')
ax.plot(mu3_linea[9:], mu4_linea[9:],color="#fca103", linewidth=6, zorder=1)
ax.scatter(mu3_crit, mu4_crit, color="#fc0303", s=300,zorder=2)
#ax.set_title("Phase transition of charge $Q_2$")
ax.set_xlabel("$\mu_3/\mu^2$")
ax.set_ylabel("$\mu_4/\mu^3$", rotation=90)
#ax.text(9, 3.2, "$(\mu_3=8, \mu_4=3)$", size=30, rotation=0.,
#         ha="center", va="center",color="white"
#         )

fig.colorbar(c, ax=ax, label='log($Q_2\mu^2$)')
fig.savefig('/Users/javier/Documents/university/8th-semester/Intro-investigacion1/codes/img.phase-trans-q2-N4.pdf')
```
